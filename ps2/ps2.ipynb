<b> Q1

```python
data = np.random.uniform(low = -100, high = 100, size= (100,100))
data
```

```python
print(f'Euclidean distance matrix :')
euclidean_dist = np.sqrt(((data[:, np.newaxis, :] - data[np.newaxis, :, :]) ** 2).sum(axis=-1))
print(euclidean_dist)
```

```python
norms = np.linalg.norm(data, axis=1, keepdims=True)
cosine_sim = np.dot(data, data.T) / (norms * norms.T)
print(cosine_sim)
```

```python
data = np.random.randint(0, 20, size=(100, 100))
data
```

```python
matches = (data[:, np.newaxis, :] == data[np.newaxis, :, :])
m = np.sum(matches, axis=-1)
p = data.shape[1]
dissimilarity_matrix = (p - m) / p
print("dissimilarity matrix")
dissimilarity_matrix
```

<b> Q3

```python
data = np.random.randint(0, 2, size=(100, 100))
data
```

```python
i_eq_1 = data[:, np.newaxis, :] == 1
j_eq_1 = data[np.newaxis, :, :] == 1

q = np.sum(i_eq_1 & j_eq_1, axis=-1)       
r = np.sum(i_eq_1 & ~j_eq_1, axis=-1)      
s = np.sum(~i_eq_1 & j_eq_1, axis=-1)      
t = np.sum(~i_eq_1 & ~j_eq_1, axis=-1)

denominator = q + r + s + t                
numerator = r + s

dissimilarity_matrix = numerator / denominator.astype(float)
print("dissimilarity_matrix")
dissimilarity_matrix
```
